### Machine Learning Pipeline

- Data Generation / Collection
  -  Data Generation - Data that is generated by us as a user, no pre existing records
  - Data Collection - The data that we collect either by surveying, regular gathering, or by web scraping / crawling some websites

- Data Cleaning
  - Cleaning any non-needed records, these can be null values, outliers that can contaminate our data
  - Even some columns that pose no usefulness is cleaned out

- Feature Engineering
  - When we have cleaned up data that we need, we start, feature engineering, ie, modifying the values as we need them
  - This can include
    - Normalising
    - Hot Encoding

- Applying Algorithms
  - After we have filtered out everything, we try to algorithmise the data, we get different trends and patterns that can be used to create the model and use it to further predict new values

- Deployment
  - The last step, after trial and testing of the algorithm, we deploy the model for normal use, which takes in the value and then gives us the predicted results

### Types of Learning:
- #### Supervised Learning
- #### Unsupervised Learning
- #### Reinforcement Learning

### Supervised Learning
- When our data is labelled


In supervised learning we basically have some attributes, and then a label, which helps us determine the values in the future

Labels can be of types:
- Continous values (Regression, 0 - 1)
- Discrete Values (Classification, Apple, Mango, Banana)

### Basic Working
1. Model Takes in Training Data
2. A hypothesis is created on the dataset
    1. Takes in some features
    2. Generates some predicted data
3. After this hypothesis is done, testing data is taken and the result for those test features is shown, and compared to actual values

### Overfitting
- This occurs when we depend **too much** on the training data, the model gives a very high accuracy on training data, but is not flexible enough and gives a very low accuracy on testing data
