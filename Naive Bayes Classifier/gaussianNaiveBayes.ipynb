### Gaussian Naive Bayes
- Used for continous features (Previous all were discrete)
- Assumed that the underlying distribution is Gaussian

For Gaussian.
$$P(x_i) = \frac{1}{\sqrt{2\pi}\sigma}e^{-\dfrac{(x - \mu)^2}{2\sigma^2}}$$

For different features, they will all be a gaussian distribution with a shifted $\mu$ and $\sigma$

#### Implementing Gaussian Naive Bayes using `SKLearn`

```python
from sklearn.naive_bayes import GaussianNB
```

```python
# For generating dataset
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("dark_background")
```

```python
gnb = GaussianNB()
```

```python
X, y = make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0, random_state=4)
X.shape, y.shape
```

#### Visualising

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.show()
```

We see this is a classifiable data

#### Training

```python
gnb.fit(X, y)
```

```python
gnb.score(X, y)
```

We get a score of $91.4\%$

```python
# Checking ourselves
ypred = gnb.predict(X)
accuracy = np.sum(ypred == y)/X.shape[0]
accuracy
```

We get a similar result
