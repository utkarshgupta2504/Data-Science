```python
import requests
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
```

```python
url = "https://www.snapdeal.com/search"
```

```python
searchParam = input("Enter search parameter: ")
```

```python
params = {
    "keyword": searchParam,
    "sort": "rlvncy",
}
```

```python
response = requests.get(url, params=params)
response.url
```

```python
response.content
```

```python
responseSoup = soup(response.content)
type(responseSoup)
```

```python
any(list(map(lambda x : "product-tuple-listing" in (x.attrs["class"] if "class" in x.attrs else []), responseSoup.findAll("div"))))
```

```python
productsList = responseSoup.findAll("div", {"class": "product-tuple-listing"})
len(productsList)
```

```python
productImages = list(map(lambda x : x.img, productsList))
len(productImages)
```

```python
productImages
```

```python
productImageSources = list(map(lambda x : x.attrs["src"] if "src" in x.attrs else x.attrs["data-src"] if "data-src" in x.attrs else None, productImages))
```

```python
productImageSources
```

```python
for pos, img in enumerate(productImageSources):
    
    response = requests.get(img)

    with open(f"./snapdealScrape/{searchParam}{pos}.jpg", "wb") as f:

        f.write(response.content)
```
