### Gradient Decent
- This concept is used to reach to the lowest error possible
- Instead of travelling in fixed steps, we travel using a gradient to reach the lowest error quicker
- Technically finds the local minima

####  Example:

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")
```

```python
x = np.arange(10)
y = (x-5)**2
x, y
```

**Goal**: We want to minimise $f(x)$

#### Visualisation:

```python
plt.plot(x, y)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

For reaching the bottom, we have to devise some way to traverse

```python
x1 = 0
learning_rate = 0.1 # The step size

# 50 steps in downhill
for i in range(50):
    gradient = 2*(x1-5) # Taking the d/dx of the function as gradient
    x1 = x1 - learning_rate * gradient
    print(x1)
```

As we can see we are very close to 5, and also at 5 the error is 0, hence our logic works

#### Visualising this:

```python
x1 = 0
learning_rate = 0.1 # The step size

plt.plot(y)
errors = []

# 50 steps in downhill
for i in range(50):
    gradient = 2*(x1-5) # Taking the d/dx of the function as gradient
    x1 = x1 - learning_rate * gradient
    error = (x1-5)**2
    errors.append(error)
    plt.scatter(x1, error)

plt.show()
```

This clearly shows how our algorithm starts with big steps, and reduces down the line, reaching the minimum error

#### Visualising the error:

```python
plt.plot(errors)
```

As can be seen, the error drops till $\sim$ 10 steps, then it saturates
