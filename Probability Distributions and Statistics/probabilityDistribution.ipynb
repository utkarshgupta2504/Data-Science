### Normal Distribution
- Depends on 2 variables
  - Variance $\sigma$
  - Mean $\mu$

- Example:
  - Lets say there are 100 students, whose mean marks is 70 and the deviation is 10
  - We get a bell curve type graph, with highest value at 70, and most of the values lying between 60 and 80

#### Generating a normal distribution and visualising

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("dark_background")
sns.set_style("dark")
```

```python
# Generating the distribution
u = 0
sigma = 1

#With this configuration, the distribution is called a _Standard_ Normal Distribution

vals = u + sigma*np.random.randn(1000) # Generates normal distribution
```

- On changing the mean $\mu$, we can shift the graph to the left or right
- On changing the std $\sigma$, we can control the spread of the graph

```python
plt.hist(vals, 50)
plt.show()
```

```python
# Plotting a bell curve
sns.histplot(vals, bins=50, kde=True)
```

We can see the bell curve has its peak at 0, the mean we provided, and max values lie between -1 and 1
