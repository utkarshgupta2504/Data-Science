## Multivariate Normal Distribution
- A univariate normal distribution was just plain old count and probability of generating a number
- Multivariate distribution deals with more than one variable value
- We plot them together to see the correlation to the final outcome

### Example - Monkeys
- Lets take 2 traits of monkeys
  - Height
  - Weight
- Hence our dataset will be $n \times 2$ matrix
- Where there are "2" features and "n" monkeys
- For m attributes / features, we get a $m \times n$ matrix

#### On plotting height x weight on a scatter plot
- We observe a definite $+ve$ correlation

- Let the means of the attributes be $\mu_x\ \&\ \mu_y$
- Let the std be $\sigma_x\ \&\ \sigma_y$
  - $\sigma_x$ can also be seen as spread in x direction
  - $\sigma_y$ can be seen as spread in y direction

Similarly for a $m$ featured dataset, There will be $m$ means and $m$ stds

$$\sum_{covariance}=\begin{pmatrix}\sigma_{xx} & \sigma_{xy} \\ \sigma_{xy} & \sigma_{yy} \end{pmatrix}$$

The probability density of 2 variables is a 3D graph with the tide in the center showing the concentration

Given:
$$X = [X_1, X_2, X_3, ..., X_m]$$
$$\mu = [\mu_1, \mu_2, \mu_3, ..., \mu_m]$$
Where $X$ and $\mu$ are column vectors,\
then,
$$P(X) = \frac{e^{\dfrac{-1}{2}(X - \mu)^T\sum^{-1}(X - \mu)}}{(2\pi)^{\frac{m}{2}}|\sum|^\frac{1}{2}}$$
Where,\
\
$\sum =>$ Covariance Matrix\
\
$|\sum| =>$ Det of Covariance
\
$m =>$ Number of dimensions


#### Bivariate Distribution using numpy

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("dark_background")
sns.set_style("dark")
```

```python
mean = np.array([0.0, 0.0])
cov = np.array([[1, 0], [0, 1]])
```

```python
dist = np.random.multivariate_normal(mean, cov, 500)
dist
```

```python
dist.shape
```

#### Plotting the points

```python
plt.scatter(dist[:, 0], dist[:, 1], alpha=0.8)
plt.show()
```

We see the center is heavily populated

```python
# Changing the cov, for more skewed diagram
cov = np.array([[1, 0.8], [0.8, 1]])
```

```python
dist = np.random.multivariate_normal(mean, cov, 500)
plt.scatter(dist[:, 0], dist[:, 1], alpha=0.8)
plt.show()
```

Since we increased the covriance, now the correlation increased, and we get a more regression like plot

#### Shifting the plot for better understanding

```python
mean2 = np.array([5.0, 6.0])
cov2 = np.array([[1, 0.2], [0.2, 1]])

dist2 = np.random.multivariate_normal(mean2, cov2, 500)
```

```python
plt.scatter(dist[:, 0], dist[:, 1])
plt.scatter(dist2[:, 0], dist2[:, 1])
plt.show()
```

We can see how the plot shifted on changing the mean, and the spread is much more owing to the std
