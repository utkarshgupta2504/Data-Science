### SKLearn
- A python library for Machine Learning
- Offers models for regressions, classifications and more

### `make_regression`
- Makes a linear regression for us

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")

%matplotlib widget
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=1000, n_features=2, n_informative=2, noise=10, random_state=1)
X.shape, y.shape
```

We just got a regression, with 2 features, both informative and with 1000 samples

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0], y)

plt.subplot(1,2,2)
plt.scatter(X[:, 1], y)

plt.show()
```

Here are the 2 features, the first may not look regression like, but in a 3D model, this can be arranged to fit in a `hyperplane`

```python
from mpl_toolkits import mplot3d

fig = plt.figure()
axes = plt.axes(projection="3d")

axes.scatter3D(X[:,0], X[:,1], y)
plt.show()
```

At azimuth = -8 and elevation = 0, we can see the regression plane

#### Implementing the regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_
```

This is the value of $\theta_1$ and $\theta_2$

```python
model.intercept_
```

This is the value of $\theta_0$

```python
model.predict([X[0], X[1]])
```

```python
y[0], y[1]
```

Pretty close

```python
model.score(X,y)
```

This model gives $\sim 99\%$ accuracy
