```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes = plt.style.available
print(themes, sep="\n")
```

```python
plt.style.use("dark_background")
```

### Line plots

```python
x = np.arange(10)

y1 = x**2
y2 = 2*x + 3

print(x, y1, y2, sep="\n")
```

```python
plt.plot(x, y1, color="red", label="Apple", marker="o")
plt.plot(x, y2, color="cyan", label="Kiwi", linestyle="dashed", marker="*")

plt.xlabel("Time")
plt.ylabel("Price")

plt.title("Prices over Time")

plt.legend()

plt.show()
```

### Scatter Plots
- Similar to line plots, but no lines joining them
- Just points floating in mid air

```python
x = np.arange(10)
y1 = x**3

plt.scatter(x, y1)
plt.show()
```

Similar test case of the line in scatter:

```python
plt.figure(figsize=(10,10)) # To enforce size

plt.scatter(x, y1, color="red", label="Apple", marker="o")
plt.scatter(x, y2, color="cyan", label="Kiwi", marker="^")

plt.xlabel("Time")
plt.ylabel("Price")

plt.title("Prices over Time")

plt.legend()

plt.show()
```

### Bar Graphs
- Mainly for comparative studies

```python
x = [0,1,2]
y = [10, 5, 15]

plt.bar(x, y)
plt.show()
```

This is a basic bar graph, for 2 inside 1:

```python
y1 = [20, 10, 12]

plt.bar(x, y1)
plt.bar(x, y)
plt.show()
```

We can shift the bar graph, using maths, and also assign the width using the `width` attribute

```python
x = np.array([0,1,2]) * 2

plt.bar(x - 0.25, y, width=0.5, label="Current Year", tick_label=["Gold", "Silver", "Copper"])
plt.bar(x + 0.25, y1, width = 0.5, label="Next year")

plt.legend()
plt.show()
```

### Pie Chart

```python
subject = "Maths", "English", "Chem", "Physics"
weightage = [20,10,15,5]

plt.pie(weightage, labels=subject)
plt.show()
```

For giving some space in between:

```python
plt.pie(weightage, labels=subject, explode=(0, 0, 0.1, 0), autopct="%1.1f%%", shadow=True)
plt.show()
```
