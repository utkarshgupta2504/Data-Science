## OpenCV
- A library to work with images in python

```python
import cv2
from matplotlib import pyplot as plt
```

Reading an image
- Using `imread`

```python
img = cv2.imread("dog.png")
img
```

As we can see, the image is rendered as a $m \times n \times 3$ matrix, storing all the `RGB` channels

```python
plt.imshow(img)
plt.show()
```

> Opencv by default assumes that the matrix is in `BGR` format, hence we get inverted Reds and Blues

We have to convert it back for a more sensible image
- Using `cvtcolor`

```python
rgbImage = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgbImage)
plt.show()
```

Now we can see the proper image of the dog
